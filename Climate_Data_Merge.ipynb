```python
import pandas as pd
files = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12', '13']
climate_df = pd.DataFrame()
for file in files:
    data = pd.read_csv(f"data/climate-monthly-{file}.csv")
    climate_df = pd.concat([climate_df, data], axis=0)
climate_df.to_csv('data/climate-monthly-all.csv', index=False)
```

```python
climate_df.info()
```

```python
climate_years_min = climate_df.groupby('STATION_NAME')['LOCAL_YEAR'].min()
climate_years_max = climate_df.groupby('STATION_NAME')['LOCAL_YEAR'].max()
climate_years = pd.merge(climate_years_min, climate_years_max, on='STATION_NAME')
```

```python
climate_years.rename(columns={'LOCAL_YEAR_x': 'First year', 'LOCAL_YEAR_y': 'Latest year'}, inplace=True)
climate_years.value_counts('Latest year')
```

```python
climate_years.loc[(climate_years['First year'] == 1976) & (climate_years['Latest year'] == 2023)]
```

```python
climate_years['Years Active'] = climate_years['Latest year'] - climate_years['First year']
```
