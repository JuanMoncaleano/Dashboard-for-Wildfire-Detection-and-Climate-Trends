```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# confirm that our new database was created
print(mongo.list_database_names())
```

```python
# assign the database to a variable name
db = mongo['wildfires']
```

```python
print(db.list_collection_names())
```

```python
print(db.climate.find_one())
print(db.fires.find_one())
```

```python
station_readings = db['climate']
fires = db['fires']
```

```python
query = {'STATION_NAME': 'PICKLE LAKE (AUT)'}
results = station_readings.find(query)
for result in results:
    print(result)
```

```python
query = {'FIRE_DISTURBANCE_AREA_IDENT': 'NIP2'}
results = fires.find(query)
for result in results:
    print(result)
```
